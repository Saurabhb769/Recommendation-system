RECOMMNEDATION SYSTEM

```python
def get_movies_titles_based_on_userid(user_id):
    movie_titles = []
    cluster = None
    for i in cluster_data['userId']:
        if cluster_data['userId'][i] == user_id:
            cluster = cluster_data['cluster'][i]
            break
    if cluster:
        for i in cluster_data['userId']:
            if cluster_data['cluster'][i] == cluster and cluster_data['title'][i] not in movie_titles:
                movie_titles.append(cluster_data['title'][i])
    else:
        print("Result not found")
    return movie_titles
user_id = int(input("Enter a user id:"))
movie_titles= get_movies_titles_based_on_userid(user_id)
if movie_titles:
    print("titles: ",movie_titles)
```
